```python
# Train/Test split (manual):

import numpy as np

# 1. Define the dataset (X)
X = np.array([[81, 72, 93],
              [68, 43, 26],
              [23, 75, 76],
              [75, 23, 24],
              [73, 12,  9]])

# 2. Create a dummy y (Labels) - must be the same length as X
y = np.array([1, 0, 1, 0, 1]) 

# 3. Calculate split (5 * 0.8 = 4)
split = int(0.8 * len(X))

# 4. Perform the slice (Ensuring X is capitalized consistently)
X_train, X_test = X[:split], X[split:]
y_train, y_test = y[:split], y[split:]

print(f"X_train rows: {len(X_train)}") # Output: 4
print(f"X_test rows: {len(X_test)}")   # Output: 1
```
