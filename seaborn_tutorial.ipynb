```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
print(sns.__version__)
iris = pd.read_csv('irisdata.csv')
iris
```

```python
iris.rename(columns= {'Sepal Length': 'sepal_len', 'Sepal Width': 'sepal_wid'}, inplace= True)
iris.rename(columns= {'Petal Length': 'petal_len', 'Petal Width': 'petal_wid','Class Label':'class_lab'}, inplace= True)
iris
```

```python
sns.scatterplot(data=iris,
                x="sepal_len",
                y="sepal_wid",
                hue="class_lab");
plt.show()
```

```python
print(iris.columns)
```

```python
sns.lineplot(data = iris, x = 'sepal_len', y = 'sepal_wid');
plt.show()
```

```python
import seaborn as sns
sns.histplot(data = iris, x = 'petal_len', y = 'petal_wid')
plt.show()
```

```python
sns.boxplot(data = iris, x = 'class_lab', y = 'petal_len')
plt.show()
```

```python
sns.violinplot(data = iris, x = 'class_lab', y = 'petal_len')
plt.show
```

```python
sns.pairplot(iris, hue='class_lab');
plt.show()
```

```python
corr = iris.corr(numeric_only= True)
sns.heatmap(corr, annot= True)
plt.show()
```

```python
sns.set_style('darkgrid')
sns.scatterplot(data = iris, x = 'sepal_len', y = 'sepal_wid', hue= 'class_lab');
plt.show()
```

```python
sns.set_style('whitegrid')
sns.scatterplot(data = iris, x = 'sepal_len', y = 'sepal_wid', hue= 'class_lab');

plt.title('Sepal Length vs Sepal Width by Class', fontsize = 16)
plt.xlabel('Sepal Length (cm)',fontsize = 12)
plt.ylabel('Sepal Width (cm)', fontsize = 12)
plt.show()
```
