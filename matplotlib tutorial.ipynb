```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
iris = pd.read_csv("irisdata.csv")
```

```python
iris
```

```python
iris.head()
```

```python
iris.rename(columns= {'Sepal Length': 'sep_len', 'Sepal Width': 'sep_wid', 'Petal Length':'pet_len', 'Petal Width': 'pet_wid', 'Class Label': 'class_lab'}, inplace = True)
```

```python
iris
```

```python
plt.scatter(iris['sep_len'], iris['sep_wid'])
plt.xlabel('Sepal Length (cm)')
plt.xlabel('Sepal Width (cm)')
plt.title('Sepal Length vs Sepal Width')
```

```python
classes = iris['class_lab'].unique()

for cls in classes:
    subset = iris[iris['class_lab'] == cls]
    plt.scatter(subset['pet_len'], subset['pet_wid'], label=cls)


plt.xlabel('Petal Length (cm)')
plt.ylabel('Petal Width (cm)')
plt.title('Petal Length vs Petal Width by Class')
plt.legend()
plt.show()
```

```python
plt.hist(iris['sep_len'])
plt.xlabel('Sepal Length')
plt.ylabel('Frequency')
plt.title('Distribution of Sepal Length')

plt.show()
```

```python
for cls in classes:
    plt.hist(
        iris[iris['class_lab'] == cls] ['pet_len'],
        bins = 10,
        alpha = 0.5,
        label = cls       
    )

plt.xlabel('Petal Length')
plt.ylabel('Frequency')
plt.title('Petal Length Distribution by Class')
plt.legend()
plt.show()
```
